## 3. 绘图属性设置

**Pyplot函数**     |  **API 方法**                         |   **描述**
* text()       |   mpl.axes.Axes.text()            |    在Axes对象的任意位置添加文字
* xlabel()     |   mpl.axes.Axes.Set_xlabel()      |    为x轴添加标签
* ylabel()     |   mpl.axes.Axes.Set_ylabel()      |    为y轴添加标签
* title()      |   mpl.axes.Axes.Set_title()       |    为Axes对象添加标题
* legend()     |   mpl.axes.Axes.legend()          |    为Axes对象添加图例
* annotate()   |   mpl.axes.Axes.annotate()        |    为Axes对象添加注释（箭头可选）
* suptitle()   |   mpl.figure。Figure.suptitle()   |    为Figure对象添加中心化标题

* 图例 legend（）

```python
plt.figure(figsize=(6,4))
x=np.linspace(0,2*np.pi)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.legend(['Sin','Cos'], #图例标签文字
          fontsize=10, #图例字号10
          loc='lower left', #图例位置左下角
           ncol=2, #图例显示为2列
           bbox_to_anchor=[0,1,1,0.2] #x坐标为0，y坐标为1，高度为1，宽度为0.2
          )
```

**loc 位置参数**
* 0 or best
* 1 or upper right
* 2 or upper left
* 3 or lower left
* 4 or lower right
* 5 or right
* 6 or center left
* 7 or center right
* 8 or lower center
* 9 or upper center
* 10 or center
* bbox_to_anchor= [x坐标,y坐标,宽度，高度]

**线条属性**
* color颜色
* linestyle样式
* linewidth宽度
* alpha透明度
* marker标记
* mfc：marker face color标记的背景色

```python
plt.figure(figsize=(6,4))
x=np.linspace(0,2*np.pi,20)
y1=np.sin(x)
y2=np.cos(x)
y3=y1-y2
# c：color颜色
# marker：标记
# ls：linestyle样式
# lw：linewidth：线宽
# ms:makersize, 标记大小
# label：标签
# mfc：marker face color：标记的背景颜色
# markersize,标记大小
# markeredgecolor: 标记边缘颜色
# markeredgewidth: 标记边缘宽度
# alpha，透明度
plt.plot(x,y1,c='r',marker='o',ls='--',lw=1,label='sin(x)',mfc='y')
plt.plot(x,y2,c='b',marker='*',ls=':',lw=1,label='cos(x)',mfc='b',markersize=10)
plt.plot(x,y3,c='g',marker='P',ls='-.',lw=1,label='sin(x)-cos(x)',mfc='r',markersize=10,alpha=0.5)
plt.legend()
```

```python
plt.figure(figsize=(5,3))
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y)

plt.title('正弦曲线',fontsize=10,color='r',loc='center')
plt.suptitle('父标题',
             y=1.1, #设置y坐标位置
             fontsize=20,
             color='b')

plt.grid(ls='--',lw=0.5,axis='y')
# ls设定网格线线条样式
# lw设置宽度
# c 设定颜色
# axis：显示特定的网格线
plt.xlabel('时间', fontsize=10, rotation=45,c='b')
plt.ylabel('幅值', fontsize=10,rotation=45, c='b')
# rotation选装角度
# fontsize设定字体大小
# color设定字体颜色
# ha 设置对齐方式水平对齐（left, center, right可选）
# va 设置对齐方式垂直对齐（bottom, center, top可选）

plt.tight_layout()
```

**文本和注释**
* text
* annotate

```python
plt.figure(figsize=(6,5))
x=np.linspace(1,10,10)
y=np.array([60,20,30,24,53,27,45,80,78,29])
plt.plot(x,y,ls='-.',marker='o',c='b')

plt.text(5,65,'I am Text',c='r',fontsize=15)
# (x坐标值，y最标值，文字内容，其它属性)

for a,b in zip(x,y):
    plt.text(a,b+2,f'{a,b}',c='b',fontsize=10,ha='center',va='bottom')

plt.annotate(
    text='线型图', #文字内容
    xy=(8,80), #坐标位置:箭头指向的位置
    xytext=(6,70),#标注内容位置:箭头开始位置
    arrowprops={ #设置箭头样式
    'width':2, #设置箭头线条宽度
        'headwidth':10, #设置箭头的宽度
        'facecolor':'r', #设置背景色
    }
)
```

### 5.2 柱形图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']='SimHei' #让图片可以显示中文
plt.rcParams['axes.unicode_minus']=False #让图片可以显示负号

plt.figure(figsize=(5,3))

x1=np.array(['语文','数学','英语','科学','社会'])
y1=np.array([82,77,65,90,59])
y2=np.array([65,70,82,77,80])
y3=np.array([70,69,75,83,49])

w=0.2
plt.bar(x1,y1,color='r',label='第一次') 
plt.bar(x1,y2,color='y',label='第二次',bottom=y1) 
plt.bar(x1,y3,color='g',label='第三次',bottom=y1+y2) 

# width设定宽度百分比
plt.xlabel('科目')
plt.ylabel('成绩')
plt.title('某班同学其中考试成绩')

for a,b,in zip(x,y1+y2+y3):
    plt.text(x=a,
            y=b,
            s=f'{b}',          
            ha='center',
             fontsize=9,)
plt.legend()
```

### 5.6 饼图

```python
x=[20,50,15,45,21,13,23]
y=['A','B','C','D','E','F','G']
plt.pie(x,
        autopct='%.1f%%',
        pctdistance=0.8,
        labels=y,
        labeldistance=1.1,
        shadow=False,
        textprops={'fontsize':10,'color':'blue',},
        explode=[0,0.1,0,0,0,0,0])
#autopct设定显示百分比，.1f表示小数点后保留1位数，%%显示百分号
#pctdistance设定百分数的位置，距离圆心的半径的百分比
# labels设定显示label
# labeldistance设定label的位置，距离圆心的半径的百分比
# shadow用于设定是否显示阴影（True or False）
# textprops设定文字样式
# explode设定某个数据是否脱离出来
plt.show()
```

```python
plt.figure(figsize=(8,8))
x=[20,50,15,45,21,13,23]
y=['A','B','C','D','E','F','G']
x2=[10,35,38,60,15,18,30]
#第一个圆环
plt.pie(x,
        autopct='%.1f%%',
        pctdistance=0.8,
        labels=y,
        labeldistance=1.1,
        textprops={'fontsize':10,'color':'black',},
        wedgeprops={ #圆环图设定参数
            'width':0.4, #设定圆环厚度
            'edgecolor':'w' #设定间隔颜色
                   })
#第二个饼图
plt.pie(x2,
        autopct='%.1f%%',
        pctdistance=0.8,
        textprops={'fontsize':10,'color':'black',},
        radius=0.6,
        )
plt.legend()
plt.show()
```

### 5.9 极坐标图

```python
N=8 #设定为8组
x=np.linspace(0,2*np.pi,N,endpoint=False) #圆周方向分为8份
height=np.random.randint(3,15,size=N) #随机生成8个数字
width=2*np.pi/N #宽度设定为圆周方向平均分成8份
color=np.random.rand(8,3) #水机生成8组3位数RGB颜色代码
axes=plt.subplot(111,projection='polar') #用子图方式生成极坐标图。projection指定为极坐标图
axes.bar(x=x,height=height,width=width,bottom=0,color=color)
```

### 5.10 雷达图

```python
plt.figure(figsize=(5,5))
N=6 #设定为6组
x=np.linspace(0,2*np.pi,N,endpoint=False) #圆周方向分为6份
y=[80,60,90,70,60,100]

#把x和y的第一个数取出，假如到原数组的最后，目的是为了让折线图能首尾连起来
x=np.concatenate((x,[x[0]]))
y=np.concatenate((y,[y[0]]))

axes=plt.subplot(111,polar=True,) #用子图方式生成极坐标图。projection指定为极坐标图
axes.plot(x,y,'o-',lw=2)
axes.fill(x,y,alpha=0.3)
axes.set_rgrids([20,40,60,80,100],fontsize=12)
```
