### 非暴力沟通的四个要素：
1. 观察
  * 清楚地表达观察的结果，而不判断或评估。
  * 不带评论的观察是人类智力的最高形式。
  * ‘每次’‘曾经’等词语表达的是观察结果。
  *  ‘经常’‘很少’这样的词语可能混淆观察和评论。
  * ‘总是’‘从不’等词语表达的是评论。
  * 如果我们表达言过其实，别人就可能产生逆反心理，而不愿做出友善的回应。

2. 感受
  * 表达正面的感受
    * 兴奋，喜悦，欣喜，甜蜜，精力充沛，兴高采烈，感激，感动，
    * 乐观，自信，振作，振奋，开心，高兴，快乐，愉快，幸福，
    * 陶醉，满足，欣慰，心旷神怡，喜出望外，平静，自在，舒适，
    * 放松，轻松，踏实，安全，温暖，放心，无忧无虑
  * 表达负面的感受
    * 害怕，担心,焦虑，忧郁，着急，紧张，心神不安，心烦意乱，忧伤，
    * 沮丧，灰心，气馁，泄气，绝望，伤感，凄凉，悲伤，恼怒，愤怒，
    * 气愤，烦恼，苦恼，生气，厌烦，不满，不快，不耐烦，不高兴，
    * 震惊，失望，困惑，茫然，寂寞，孤独，郁闷，难过，悲观，沉重，
    * 麻木，精疲力尽，萎靡不振，疲惫不堪，昏昏欲睡，无精打采，尴尬，
    * 惭愧，内疚，嫉妒，遗憾，不舒服，失落，郁闷，烦，等等。
  * 区分‘想法’与‘感受’
      * “我觉得”常常是在表达想法，而不是感受。
          * 我觉得你应该懂得更多
          * 我觉得自己很无能
          * 我觉得老板很卑鄙
          * 我觉得他很负责任
          * 被抛弃、被欺负、无人理睬、得不到支持、无人赏识、被利用、被贬低
  * 感受的根源
    * 感受根源于自身。我们的需要和期待，对他人的言行的看法，导致了我们的感受。
    * 听到不中听的话的四种选择
      1. 对内：认为自己犯了错
      2. 对外：指责对方
      3. 了解自己的感受和需要
      4. 体会他人的感受和需要
  * 表达感受
    * 我（感到）。。。因为我。。。

3. 需要
  * 说出哪些需要导致那样的感受。
  * 一旦人们开始谈论需要，而不是指责对方，他们就有可能找到办法来满足对方的需求。
  * 每个人都有的基本需求：
    * 自由选择（梦想，目标，方向）
    * 庆祝（生命，创造力，梦想成真，纪念人生的失落，亲人去世，梦想破灭，表达悲伤等）
    * 言行一致（真诚，创造，意义，自我肯定）
    * 滋养身体（空气，食物，运动，免于病毒、细菌、昆虫及肉食动物伤害，休息，住所，触摸，水）
    * 玩耍（乐趣，欢笑）
    * 情意相通（美，和谐，激励，秩序，平静）
    * 相互依存（接纳，欣赏，亲密关系，社区，体贴，成长，安全感，倾听，诚实，爱，信心，尊重，支持，信任，理解）

4. 请求
  * 说出具体要求。清楚告诉对方，我们希望他们做什么。
  * 如果我们请求他们不做什么，对方也许会很困惑，不知道我们到底想要什么。而且，这样的请求还易引起反感。
  * 确认对方是否明白
  * 当对方给予反馈，表达我们的感激
  * 当对方不愿反馈，倾听他们的感受和需要
  * 询问对方此时从此刻的感受
    * 听我说这些，你的心情怎么样？为什么呢？
  * 询问对方正在想什么
    * 请你谈谈我的想法是否具有可行性？入果不行，根据你的判断，哪些因素妨碍建议的实施？
  * 询问对方是否接受我们的请求
    * 你是否同意将会议推迟一周？

`
举例：
母亲对儿子说：Felix，看到咖啡桌下的袜子[观察]，我不太高兴[感受]，因为我看重整洁[需要],你是否愿意将袜子拿到房间，放进洗衣机？[请求]。
`

> 在道德与不道德的区分之外，有片田野。我将在那里见你。-lumi  
> 语言使我们陷于是非之中。他擅长将人分类，好人或坏人，正常或不正常，负责任或不负责任，聪明或愚蠢...  
> 貌似客观的于洋表达自己，一旦遇到不喜欢的人或不理解的事，就会想：别人呢有什么不对。
* 比较
* 强人所难（命令和指责）
  1. 我们对别人的要求，往往暗含着威胁，如果不配合，他们就会受到惩罚。这是强者常用的手段。
  2. 区别请求与命令
     * 请求没有得到满足时，提出请求的人如果批评和指责，那就是命令。
     * 如果利用对方的内疚来达到目的，也是命令。
     * 在提出请求前，如果我们有以下想法，那么我们的请求已经成为对他人的要求。
       * 他应该把房间整理干净
       * 他必须照我说的去做
       * 老板应当给我提薪
       * 我有权要求更长的假期
  3. 有些行为值得奖励，而另一些行为必须受罚。通过‘应当’，将行为分为好坏，人分为好坏。
  4. 人的行为应该是追求幸福，而不是逃避惩罚。

> '不应该'，'应该'，'不得不'，这些表达方式特别适合专制目的，人们越是习惯于评定是非，他们也就越倾向于追随权威，来获得正确错误的标准。一旦专注于自身的感受和需要，我们就不再是好奴隶或好下属。

__当我们根据以下理由行动时，我们就是在试图**回避责任**__
  1. 受数不清楚的力量的驱使
     * _为什么打扫房间？因为我不得不做_
  2. 我们的个人情况，成长历程，自我形象等
     * _为什么喝酒？因为我是酒鬼_
  3. 其它人的行为  
     * _为什么要打自己的孩子？因为他不好好学习_
  4. 上级的命令  
     * _为什么欺骗顾客？因为老板让我这么做_
  5. 同伴的压力  
     * _为什么抽样？因为同伴都抽烟_
  6. 机构的规章制度及政策  
     * _为什么停职？因为你违规了，我必须这么做_
  7. 性别角色，社会角色或年龄角色  
     * _为什么我必须做我讨厌的工作？因为我不仅是丈夫，还是孩子的父亲_
  8. 无法控制的冲动  
     * _为什么吃糖？因为吃糖的冲动控制了我_

> 人类的灭亡是因为现代人的唯唯诺诺，缺乏责任感，毕恭毕敬地服从命令。

## 第七章 用全身心倾听

> 不要着急做什么，站在那里

* 如果一个人想要别人了解他的处境，听到的却是安慰和建议，那么，他就可能觉得不舒服。
* 哪些行为会妨碍我们体会他人的处境
  * 建议：‘我想你应该。。。’
  * 比较：‘这算不了什么。你听听我的经历。。。’
  * 说教：‘如果你这样做。。。你将会得到很大的好处。’
  * 安慰：‘这不是你的错，你已经尽最大努力了。’
  * 回忆：‘这让我想起。。。’
  * 否定：‘高兴一点，不要这么难过’
  * 同情：‘哦，你真可怜’
  * 询问：‘这种情况是什么时候开始的？’
  * 辩解：‘我本想早点打电话給你，但昨晚。。。’
  * 纠正：‘事情的经过不是这样的’
 
* 分析妨碍了倾听
* 急于行动使我们无法充分体会他们的状况。
* 不论别人用什么方式表达自己，我们都可以用心体会其中所包含的观察，感受，需要和请求。
* 当别人指着你时，你要弄明白他需要什么，而不是反击。

* 当对方得到充分的倾听
  * 我们将会体会到气氛变得轻松
  * 他会停止谈话

* 大声地提出请求
* 换一个环境。我们需要时间和空间来调整状态，等待平静下来，再回来。

## 第十章 充分表达愤怒

> 我想说的是，杀人太肤浅了。再生气时，杀人，打人，骂人都无法真正传达我的心声。如果真的很生气，我们就需要强有力的方式充分表达自己。

> 如果在一个社会，内疚被运用来控制人。那么，指责他人就容易成为一种习惯。此时我们认为另一个人会主导我们的情绪。

> 如果我们把人看作‘正直的’和‘不正直的’，‘负责人的’和‘不负责任的’，那么我们的想法也会带来暴力。

  * 非暴力沟通不主张忽视或压抑愤怒。通过深入分析愤怒，可以充分表达内心渴望。
  * 我们为什么生气
    * 充分表达愤怒的第一步是我们不再归咎于他人。
    * 专注于我们的需要，必批评他人是什么人，更有益于生活。
    * 愤怒是我们的思维方式造成的。他的核心是尚未满足的需求。
    * 我生气是因为我需要......
  * 表达愤怒的四个步骤
    1. 停下来，除了深呼吸，什么都别做。避免采取行动去指责或惩罚对方。
    2. 想一想什么想法使自己生气了
    3. 体会自己想要满足的需要
      * 先倾听他人，表达我们的理解
      * 给自己时间
    4. 表达我们的感受，说出我们的需求
  * 只要人们人们认为自己受到了指责，他们就很难体会到别人的痛苦。

## 第十一章 运用强制力避免伤害

> 在非暴力沟通中，运用强制力是出于防卫的目的，而不是为了惩罚对方。抱住年幼的孩子不让他跑到公路上是为了他的安全，如果打他屁股，就是在惩罚他。

  * 在使用防卫性强制时，我们的注意力在自己和他人的需求上，而不评价对方及其行为。
  * 什么叫不成熟：
    * 意识不到自己行为的后果
    * 认识不到并不需要通过惩罚来满足自己的需要
    * 相信有权力去惩罚或伤害他人，或他人罪有应得
    * 产生了幻觉
  * 在使用惩罚性强制时，往往加强了对方的敌意和抵触心理，使双方关系更加疏远。
  * 惩罚的类型：
    * 体罚：对体罚的恐惧会让孩子难以体会的父母的爱。
    * 指责
    * 否定他人
    * 不给孩子某种好处
    * 变得冷漠是一种强有力的威胁
  * 惩罚的代价
    * 为了回避惩罚，忽视事情本身的价值。陷于对失败的忧虑中。
    * 自尊受损
    * 关系疏远
    * **_【一旦我们被看作施暴的人，我们就很难得到友善的回应】_**
  * 惩罚的局限性
    * “我希望他基于怎样的原因去做我想要他去做的事情？”
    * 惩罚或奖励很难实现我们的目的。

## 第十三章 表达感激

> 赞扬并不总是有助于深化彼此的联系。在赞扬他人时，我们很少揭示内心活动，而把自己放在了裁判的位置。

> 不要那么谦虚，因为你没那么伟大。

  * 一旦员工意识到经理的赞扬是为了操纵他们，就可能产生逆反心理。而发现这些赞扬潜在的目的后，对于由衷的感激也会心存疑虑。
  * 如果利用赞扬来施加影响，对方还可能误解我们的意思。
  * 赞扬只是为了庆祝他人的行为提升了我们的生活品质，而不是想得到任何回报。
  * 表达感激的方式：
    * 对方做了什么事情使得我们的生活得到了改善
    * 我们的哪些需求得到了满足
    * 我们的心情怎么样
  * 接受别人的感激

* 儿子不敢问我要5块钱，我给了他，并提醒他以后要问我要钱，不要问爷爷要。他任然不相信我会给他钱。
* 儿子能想着爷爷，说明他心里关心爷爷，不是一个自私的人。
